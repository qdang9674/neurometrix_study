```python
import pandas as pd
import psycopg2
import time
import numpy as np
import re
from datetime import timedelta
import pickle
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt
from scipy import stats
import neurokit2 as nk
import os
```

```python
LOAD_DIR = r'C:\Users\brook\Quang_workstation\Shooting_game\Extract_data'
# Define SAVE_DIR
SAVE_DIR = r'C:\Users\brook\Quang_workstation\Shooting_game\Output\output_2'

```

# Load data

```python
cluster_df = pd.read_csv(LOAD_DIR + '\cluster_april_22.csv')
cluster_df = cluster_df.rename({'subject_id' : 'Subject_id'},axis=1)
cluster_df
```

```python
data_df = pd.read_csv(LOAD_DIR + '\shooting_game_data_2024_4_25.csv')
data_df = pd.merge(data_df,cluster_df,on='Subject_id', how='left')

data_df
```

```python
#add team_type
rows = []

for team_number in data_df['Team_number'].unique():
    team_data = data_df[data_df['Team_number'] == team_number]
    cluster_labels = team_data['cluster_labels'].unique()
    team_type = np.nan
    
    # Determine the team type based on cluster labels
    if len(cluster_labels) == 1:  # Both team members belong to the same cluster
        if cluster_labels[0] == 1:
            team_type = '(1,1)'
        elif cluster_labels[0] == 2:
            team_type = '(2,2)'
        elif cluster_labels[0] == 3:
            team_type = '(3,3)'
    else:  # Team members belong to different clusters
        if 1 in cluster_labels and 2 in cluster_labels:
            team_type = '(1,2)'
        elif 1 in cluster_labels and 3 in cluster_labels:
            team_type = '(1,3)'
        elif 2 in cluster_labels and 3 in cluster_labels:
            team_type = '(2,3)'

    rows.append([team_number,team_type])
                
                
temp_df = pd.DataFrame(rows,columns=['Team_number','team_type'])
temp_df
```

```python
# Merge the team_types_df back to the original data_df based on Team_number
data_df = data_df.merge(temp_df, on='Team_number',how ='left')
data_df
```

# Ploting

```python
# Filter data for winning and losing teams
win_data = data_df[data_df['screenshot_result'] == 'win']
lose_data = data_df[data_df['screenshot_result'] == 'lost']

# Calculate counts and percentages for entire, winning, and losing teams
entire_cluster_counts = data_df['cluster_labels'].value_counts()
win_cluster_counts = win_data['cluster_labels'].value_counts()
lose_cluster_counts = lose_data['cluster_labels'].value_counts()

entire_percentages = entire_cluster_counts / entire_cluster_counts.sum() * 100
win_percentages = win_cluster_counts / win_cluster_counts.sum() * 100
lose_percentages = lose_cluster_counts / lose_cluster_counts.sum() * 100

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Entire distribution
axes[0].pie(entire_percentages, labels=entire_percentages.index, autopct='%1.1f%%', startangle=140)
axes[0].set_title('Entire Distribution')

# Winning team distribution
axes[1].pie(win_percentages, labels=win_percentages.index, autopct='%1.1f%%', startangle=140)
axes[1].set_title('Winning Team Distribution')

# Losing team distribution
axes[2].pie(lose_percentages, labels=lose_percentages.index, autopct='%1.1f%%', startangle=140)
axes[2].set_title('Losing Team Distribution')

# Equal aspect ratio ensures that pie is drawn as a circle.
for ax in axes:
    ax.axis('equal')

# Ensure SAVE_DIR exists or create it
if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)
plt.savefig(SAVE_DIR + '\cluster_distribution.png', dpi=500)

plt.show()
```

```python
# Define the columns for subplots
columns = ['Q5_trusted_by_teamate', 'Q7_trusted_by_teamate', 'Q10_trusted_by_teamate']
Question_name = {
    'Q5_trusted_by_teamate': 'Trusted by teammate: 1-20',
    'Q7_trusted_by_teamate': 'Rely by teammate: 0-100',
    'Q10_trusted_by_teamate': 'Confidence by teammate 0-100',
}

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Iterate over each column and create a bar chart in each subplot
for i, col in enumerate(columns):
    mean_trusted_by_teamate = data_df.groupby('cluster_labels')[col].mean()
    axes[i].bar(mean_trusted_by_teamate.index, mean_trusted_by_teamate.values)
    axes[i].set_xlabel('Cluster Labels')
    axes[i].set_xticks([1,2,3])
    axes[i].set_ylabel(f'Mean score')
    axes[i].set_title(f'Mean of score for: {Question_name[col]}')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability if needed
    #axes[i].grid(axis='y')  # Add gridlines along the y-axis
plt.tight_layout()  # Adjust layout to prevent overlapping

# Ensure SAVE_DIR exists or create it
if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)
plt.savefig(SAVE_DIR + '\mean_trust_by_teammate.png', dpi=500)

plt.show()
```

# Ploting on team number

```python

# Filter data for winning and losing teams
win_data = data_df[data_df['screenshot_result'] == 'win']
lose_data = data_df[data_df['screenshot_result'] == 'lost']

# Calculate counts and percentages for entire, winning, and losing teams
entire_cluster_counts = data_df['team_type'].value_counts()
win_cluster_counts = win_data['team_type'].value_counts()
lose_cluster_counts = lose_data['team_type'].value_counts()


# Merge counts into a single DataFrame
merged_counts = pd.DataFrame({
    'Entire': entire_cluster_counts,
    'Winning': win_cluster_counts,
    'Losing': lose_cluster_counts
}).fillna(0)  # Fill NaN values with 0

# Calculate percentages
merged_percentages = merged_counts.div(merged_counts.sum(axis=0), axis=1) * 100

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot pie charts for entire, winning, and losing teams
for i, (title, percentages) in enumerate(merged_percentages.items()):
    axes[i].pie(percentages, labels=percentages.index, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(title + ' Team Distribution')
    axes[i].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Ensure SAVE_DIR exists or create it
if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)

# Save the figure
plt.savefig(SAVE_DIR+ r'\team_type_distribution.png', dpi=500)

plt.show()
```
