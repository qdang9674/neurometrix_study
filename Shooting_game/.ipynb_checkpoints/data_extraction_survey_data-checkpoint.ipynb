```python
def sum_damage_weapon(data, isGun):
    sum = 0
    for i in data:
        if isGun == True:
            if i == 15:
                sum += i
        if isGun == False:
            if i != 15:
                sum += i
    return sum


def size_damage_weapon(data, isGun):
    size = 0
    for i in data:
        if isGun == True:
            if i == 15:
                size += 1
        if isGun == False:
            if i != 15:
                size += 1
    return size

```

```python
def calculate_reaction_time1(gunshots_df, silence_time, time_window):
    """
    Calculates the reaction time between t1 and t2 based on specified conditions.
    Args:
        df_result (pandas.DataFrame): The input DataFrame containing the data.
        silence_time (int): The silence time in seconds after t2 and before t1. We use same silence time for two cases
        time_window (int): The time window in seconds to consider t2 after t1.
    Returns:
        pandas.DataFrame: The modified DataFrame with the reaction time calculated and updated.
    """
    HITER_NAME = 'damager_name' #column of hitter name
    VICTIM_NAME = 'damaged_name' #column of victim name
    gunshots_df_copy = gunshots_df.copy()
    gunshots_df_copy = gunshots_df_copy.sort_values('utc').reset_index(drop=True)
    gunshots_df_copy['utc'] = pd.to_datetime(gunshots_df_copy['utc'])

    # Define the pattern to match t1_victim_actor_name
    pattern = r'(mammal|reptile)'


    # Create a list to store the indices of dropped rows
    dropped_indices = []
    gunshots_df_copy['RT_1'] = np.nan
    for index, row in gunshots_df_copy.iterrows():
        # Check if the current row index is in the dropped_indices list
        if index in dropped_indices:
            continue  # Skip to the next iteration

        # Retrieve the values from the current row
        t1_victim_actor_name = row[VICTIM_NAME]
        t1_attacking_vehicle = row[HITER_NAME]
        # Check if t1_victim_actor_name does not match the pattern and there are no shots within the previous silence time
        if (
            not pd.isna(t1_victim_actor_name)
            and re.match(pattern, t1_victim_actor_name)
            and not (
                (gunshots_df_copy[VICTIM_NAME] == t1_attacking_vehicle) &
                (gunshots_df_copy[HITER_NAME] == t1_victim_actor_name) &
                (gunshots_df_copy['utc'] > row["utc"] - timedelta(seconds=silence_time)) &
                (gunshots_df_copy['utc'] < row["utc"])
            ).any()
        ):
            # Find the t2 row where shot_name matches t1's shooter_name
            t2_row = gunshots_df_copy[
                (gunshots_df_copy[VICTIM_NAME] == t1_attacking_vehicle) &
                (gunshots_df_copy[HITER_NAME] == t1_victim_actor_name) &
                (gunshots_df_copy.index > index)  # Ensure t2 rows are after t1
            ]

            if not t2_row.empty:
                # Retrieve the timestamps of t1 and t2
                t1_timestamp = gunshots_df_copy.loc[index, 'utc']
                t2_timestamp = t2_row.iloc[0]['utc']

                # Calculate the reaction time
                reaction_time = t2_timestamp - t1_timestamp

                # Check if reaction_time is lower than the specified time window
                if reaction_time.total_seconds() < time_window:
                    # Define the start and end timestamps for the time window with silence time
                    start_time = t1_timestamp
                    end_time = t2_timestamp + timedelta(seconds=silence_time)

                    # Mask duplicate rows with the same t1_victim_actor_name and t1_attacking_vehicle between t1 and t2 plus silence_time
                    mask = (
                        (
                            (gunshots_df_copy[VICTIM_NAME] == t1_victim_actor_name) &
                            (gunshots_df_copy[HITER_NAME] == t1_attacking_vehicle)
                        ) |
                        (
                            (gunshots_df_copy[HITER_NAME] == t1_victim_actor_name) &
                            (gunshots_df_copy[VICTIM_NAME] == t1_attacking_vehicle)
                        )
                    ) & (
                        (gunshots_df_copy['utc'] > start_time) &
                        (gunshots_df_copy['utc'] < end_time)
                    )
                    dropped_indices.extend(gunshots_df_copy[mask].index)  # Add the indices of dropped rows to the list

                    # Add the reaction time to the corresponding t1 row
                    gunshots_df_copy.at[index, 'RT_1'] = reaction_time.total_seconds()
                    gunshots_df_copy.at[t2_row.iloc[0].name, 'RT_1'] = reaction_time.total_seconds()

        filtered_df1 = gunshots_df_copy[~gunshots_df_copy['RT_1'].isna()]
        filtered_df1 = filtered_df1.sort_values(by='RT_1')
        grouped_df = filtered_df1.groupby([HITER_NAME,VICTIM_NAME, 'RT_1']).size().reset_index().drop(columns=0)
        grouped_df = grouped_df[(grouped_df["RT_1"] > 0.1) & (grouped_df["RT_1"] < (time_window - 2))]
        grouped_df = grouped_df[~grouped_df['RT_1'].duplicated(keep='first')]
        filtered_df = grouped_df[grouped_df[HITER_NAME].str.extract(pattern, expand=False).notna()]

    return filtered_df1, filtered_df
```

```python
#calculate ECG data function

def get_ecg_features(ecg, time_in_sec, fs):
    """
    Compute ECG features from raw ECG signal.

    Parameters
    ----------
    ecg : array-like
        Raw ECG signal.
    time_in_sec : array-like
        Timestamps corresponding to each sample of the ECG signal.
    fs : float
        Sampling frequency of the ECG signal.

    Returns
    -------
    array
        Array of ECG features: [mean heart rate, maximum heart rate, minimum heart rate, heart rate variability].
    """
    try:
        b, a = butter(4, (0.25, 25), 'bandpass', fs=fs)
        ecg_filt = filtfilt(b, a, ecg, axis=0)
        ecg_cleaned = nk.ecg_clean(ecg_filt, sampling_rate=fs)
        instant_peaks, rpeaks = nk.ecg_peaks(ecg_cleaned, sampling_rate=fs,method="engzeemod2012")
    except Exception as e:
        raise ValueError("Error processing ECG signal: " + str(e))

    rr_times = time_in_sec[rpeaks['ECG_R_Peaks']]
    if len(rr_times) == 0:
        raise ValueError("No R-peaks detected in ECG signal.")
    
    # Assuming d_rr contains the time intervals between successive heartbeats in seconds
    d_rr = np.diff(rr_times)
    heart_rate = 60 / d_rr
    if heart_rate.size == 0:
        raise ValueError("Error computing heart rate from ECG signal.")
    
    valid_heart_rate = heart_rate[~np.isnan(heart_rate)]
    z_scores = np.abs(stats.zscore(valid_heart_rate))

    # Define a z-score threshold beyond which a value is considered an outlier
    z_score_threshold = 2.0

    # Remove outliers from the valid_heart_rate array
    heart_rate = valid_heart_rate[z_scores <= z_score_threshold]

    hr_mean = np.nanmean(heart_rate)
    hr_min = np.nanmin(heart_rate)
    hr_max = np.nanmax(heart_rate)
    d_rr_ms = 1000 * d_rr
    d_d_rr_ms = np.diff(d_rr_ms)

    valid_d_d_rr_ms = d_d_rr_ms[~np.isnan(d_d_rr_ms)] 
    z_scores = np.abs(stats.zscore(valid_d_d_rr_ms))
    d_d_rr_ms= valid_d_d_rr_ms[z_scores <= z_score_threshold]
    heart_rate_variability = np.sqrt(np.nanmean(np.square(d_d_rr_ms)))

    # Create a new signal 'ecg_with_rr_intervals' with RR intervals and a 1-second window around each RR interval
    ecg_with_rr_intervals = []
    ecg_with_rr_intervals_cleaned = []

    for rr_interval in rr_times:
        start_time = rr_interval - 0.1 # 1 second before the RR interval
        end_time = rr_interval + 0.1   # 1 second after the RR interval
        indices = np.where((time_in_sec >= start_time) & (time_in_sec <= end_time))[0]

        # Validate indices to ensure they are within bounds
        indices = indices[(indices >= 0) & (indices < len(ecg))]

        if len(indices) > 0:
            ecg_with_rr_intervals.extend(ecg[indices])
            ecg_with_rr_intervals_cleaned.extend(ecg_cleaned[indices])

    # Convert lists to NumPy arrays
    ecg_with_rr_intervals = np.array(ecg_with_rr_intervals)
    ecg_with_rr_intervals_cleaned = np.array(ecg_with_rr_intervals_cleaned)

    # Calculate noise power (mean squared amplitude of noise)
    signal_power = np.var(ecg_with_rr_intervals)
    noise_power = np.var(ecg_with_rr_intervals - ecg_with_rr_intervals_cleaned)

    # Calculate noise power (mean squared amplitude of noise)
    #signal_power = np.var(ecg)
    #noise_power = np.var(ecg - ecg_cleaned)

     # Calculate SNR in dB and append it to the array
    snr_values = 10 * np.log10(signal_power / noise_power)
    
    return hr_mean, hr_max, hr_min, heart_rate_variability, snr_values
```
