```python
def calculate_reaction_time1(gunshots_df, silence_time, time_window):
    """
    Calculates the reaction time between t1 and t2 based on specified conditions.
    Args:
        df_result (pandas.DataFrame): The input DataFrame containing the data.
        silence_time (int): The silence time in seconds after t2 and before t1. We use same silence time for two cases
        time_window (int): The time window in seconds to consider t2 after t1.
    Returns:
        pandas.DataFrame: The modified DataFrame with the reaction time calculated and updated.
    """

    gunshots_df_copy = gunshots_df.copy()
    gunshots_df_copy = gunshots_df_copy.sort_values('utc').reset_index(drop=True)
    gunshots_df_copy['utc'] = pd.to_datetime(gunshots_df_copy['utc'])

    # Define the pattern to match t1_victim_actor_name
    pattern = r'(mammal|reptile)'


    # Create a list to store the indices of dropped rows
    dropped_indices = []
    gunshots_df_copy['RT_1'] = np.nan
    for index, row in gunshots_df_copy.iterrows():
        # Check if the current row index is in the dropped_indices list
        if index in dropped_indices:
            continue  # Skip to the next iteration

        # Retrieve the values from the current row
        t1_victim_actor_name = row['shot_name']
        t1_attacking_vehicle = row['shooter_name']
        # Check if t1_victim_actor_name does not match the pattern and there are no shots within the previous silence time
        if (
            not pd.isna(t1_victim_actor_name)
            and re.match(pattern, t1_victim_actor_name)
            and not (
                (gunshots_df_copy['shot_name'] == t1_attacking_vehicle) &
                (gunshots_df_copy['shooter_name'] == t1_victim_actor_name) &
                (gunshots_df_copy['utc'] > row["utc"] - timedelta(seconds=silence_time)) &
                (gunshots_df_copy['utc'] < row["utc"])
            ).any()
        ):
            # Find the t2 row where shot_name matches t1's shooter_name
            t2_row = gunshots_df_copy[
                (gunshots_df_copy['shot_name'] == t1_attacking_vehicle) &
                (gunshots_df_copy['shooter_name'] == t1_victim_actor_name) &
                (gunshots_df_copy.index > index)  # Ensure t2 rows are after t1
            ]

            if not t2_row.empty:
                # Retrieve the timestamps of t1 and t2
                t1_timestamp = gunshots_df_copy.loc[index, 'utc']
                t2_timestamp = t2_row.iloc[0]['utc']

                # Calculate the reaction time
                reaction_time = t2_timestamp - t1_timestamp

                # Check if reaction_time is lower than the specified time window
                if reaction_time.total_seconds() < time_window:
                    # Define the start and end timestamps for the time window with silence time
                    start_time = t1_timestamp
                    end_time = t2_timestamp + timedelta(seconds=silence_time)

                    # Mask duplicate rows with the same t1_victim_actor_name and t1_attacking_vehicle between t1 and t2 plus silence_time
                    mask = (
                        (
                            (gunshots_df_copy['shot_name'] == t1_victim_actor_name) &
                            (gunshots_df_copy['shooter_name'] == t1_attacking_vehicle)
                        ) |
                        (
                            (gunshots_df_copy['shooter_name'] == t1_victim_actor_name) &
                            (gunshots_df_copy['shot_name'] == t1_attacking_vehicle)
                        )
                    ) & (
                        (gunshots_df_copy['utc'] > start_time) &
                        (gunshots_df_copy['utc'] < end_time)
                    )
                    dropped_indices.extend(gunshots_df_copy[mask].index)  # Add the indices of dropped rows to the list

                    # Add the reaction time to the corresponding t1 row
                    gunshots_df_copy.at[index, 'RT_1'] = reaction_time.total_seconds()
                    gunshots_df_copy.at[t2_row.iloc[0].name, 'RT_1'] = reaction_time.total_seconds()

        filtered_df1 = gunshots_df_copy[~gunshots_df_copy['RT_1'].isna()]
        filtered_df1 = filtered_df1.sort_values(by='RT_1')
        grouped_df = filtered_df1.groupby(['shooter_name','shot_name', 'RT_1']).size().reset_index().drop(columns=0)
        grouped_df = grouped_df[(grouped_df["RT_1"] > 0.1) & (grouped_df["RT_1"] < (time_window - 2))]
        grouped_df = grouped_df[~grouped_df['RT_1'].duplicated(keep='first')]
        filtered_df = grouped_df[grouped_df['shooter_name'].str.extract(pattern, expand=False).notna()]

    return filtered_df1, filtered_df
```

```python
import pandas as pd
import numpy as np
from datetime import timedelta
import re

def calculate_reaction_time2(gunshots_df, silence_time, time_window):
    """
    Calculates the reaction time between t1 and t2 based on specified conditions.
    Args:
        gunshots_df (pandas.DataFrame): The input DataFrame containing the data.
        silence_time (int): The silence time in seconds after t2 and before t1. We use same silence time for two cases
        time_window (int): The time window in seconds to consider t2 after t1.
    Returns:
        pandas.DataFrame: The modified DataFrame with the reaction time calculated and updated.
    """

    gunshots_df_copy = gunshots_df.copy()
    gunshots_df_copy = gunshots_df_copy.sort_values('utc').reset_index(drop=True)
    gunshots_df_copy['utc'] = pd.to_datetime(gunshots_df_copy['utc'])

    # Define the pattern to match t1_victim_actor_name
    pattern = r'(mammal|reptile)'

    # Create a list to store the indices of dropped rows
    dropped_indices = []
    gunshots_df_copy['RT_2'] = np.nan
    for index, row in gunshots_df_copy.iterrows():
        # Check if the current row index is in the dropped_indices list
        if index in dropped_indices:
            continue  # Skip to the next iteration

        # Retrieve the values from the current row
        t1_victim_actor_name = row['shot_name']
        t1_attacking_vehicle = row['shooter_name']
        # Check if t1_victim_actor_name does not match the pattern and there are no shots within the previous silence time
        if (
            not pd.isna(t1_victim_actor_name)
            and re.match(pattern, t1_victim_actor_name)
            and not (
                (gunshots_df_copy['shot_name'] == t1_attacking_vehicle) &
                (gunshots_df_copy['shooter_name'] == t1_victim_actor_name) &
                (gunshots_df_copy['utc'] > row["utc"] - timedelta(seconds=silence_time)) &
                (gunshots_df_copy['utc'] < row["utc"])
            ).any()
        ):
            # Find the t2 row where the shooter name matches t1_victim_actor_name
            t2_row = gunshots_df_copy[
                (gunshots_df_copy['shooter_name'] == t1_victim_actor_name) &
                (gunshots_df_copy.index > index)  # Ensure t2 rows are after t1
            ]

            if not t2_row.empty:
                # Retrieve the timestamps of t1 and t2
                t1_timestamp = gunshots_df_copy.loc[index, 'utc']
                t2_timestamp = t2_row.iloc[0]['utc']

                # Calculate the reaction time
                reaction_time = t2_timestamp - t1_timestamp

                # Check if reaction_time is lower than the specified time window
                if reaction_time.total_seconds() < time_window:
                    # Define the start and end timestamps for the time window with silence time
                    start_time = t1_timestamp
                    end_time = t2_timestamp + timedelta(seconds=silence_time)

                    # Mask duplicate rows with the same t1_victim_actor_name and t1_attacking_vehicle between t1 and t2 plus silence_time
                    mask = (
                        (
                            (gunshots_df_copy['shot_name'] == t1_victim_actor_name) &
                            (gunshots_df_copy['shooter_name'] == t1_attacking_vehicle)
                        ) |
                        (
                            (gunshots_df_copy['shooter_name'] == t1_victim_actor_name)
                        )
                    ) & (
                        (gunshots_df_copy['utc'] > start_time) &
                        (gunshots_df_copy['utc'] < end_time)
                    )
                    dropped_indices.extend(gunshots_df_copy[mask].index)  # Add the indices of dropped rows to the list

                    # Add the reaction time to the corresponding t1 row
                    gunshots_df_copy.at[index, 'RT_2'] = reaction_time.total_seconds()
                    gunshots_df_copy.at[t2_row.iloc[0].name, 'RT_2'] = reaction_time.total_seconds()

    filtered_df1 = gunshots_df_copy[~gunshots_df_copy['RT_2'].isna()]
    filtered_df1 = filtered_df1.sort_values(by='RT_2')
    grouped_df = filtered_df1.groupby(['shooter_name','shot_name', 'RT_2']).size().reset_index().drop(columns=0)
    grouped_df = grouped_df[(grouped_df["RT_2"] > 0.1) & (grouped_df["RT_2"] < (time_window - 2))]
    filtered_df = grouped_df[grouped_df['shooter_name'].str.extract(pattern, expand=False).notna()]

    return  filtered_df1, filtered_df 
```
