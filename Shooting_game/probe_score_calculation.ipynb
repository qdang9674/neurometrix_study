# Connect database and load df

```python
DB_NAME = "cognitive_battery"

conn = psycopg2.connect(
    host="127.0.0.1",
    port="5432",
    database=DB_NAME,
    user="postgres",
    password="postgres"
)
```

```python
# Execute the SQL query
cursor = conn.cursor()
sql_query = "SELECT * FROM public.survey_data"
cursor.execute(sql_query)

# Fetch all the results
columns = [desc[0] for desc in cursor.description]
data = cursor.fetchall()
df_survey = pd.DataFrame(data, columns=columns)

cursor.close()
#eye_df
```

```python
df_survey
```

**df_demographic**

```python
# Function to map education levels
def map_education_level(education):
    if 'High school' in education:
        return 'GED'
    elif 'Associate' in education or 'Bachelor' in education:
        return 'Undergraduate'
    elif 'Master' in education:
        return 'Graduate'
    else:
        return education


df_demographic = pd.DataFrame(df_survey['demographic_inventory_data'].to_list())
df_demographic = df_demographic.applymap(lambda x: x[0])  #remove list
df_demographic = pd.concat([df_survey[['userid']],df_demographic],axis=1) #add userid
df_demographic = df_demographic[['userid','Q2','Q3','Q4','Q6']].rename({'Q2':'Age','Q3':'Gender','Q4':'Race','Q6':'Education',},axis=1) #filter and remane columns
df_demographic['Education'] = df_demographic['Education'].apply(map_education_level)


df_demographic
```

**Calculate trait_score and state_score**

```python
REVERSE_QUESTIONS_OLD = [
    'Q2', 'Q3', 'Q6', 'Q9', 'Q12', 'Q16', 'Q17', 'Q20', 'Q21', 'Q22', 'Q23', 'Q26', 'Q29', 'Q32', 'Q36', 'Q37', 'Q40', 'Q41'
]
REVERSE_QUESTIONS_NEW = [
    'Q3', 'Q5', 'Q6', 'Q9', 'Q10', 'Q12', 'Q13', 'Q16', 'Q18', 'Q19', 'Q21', 'Q22', 'Q23', 'Q26', 'Q29', 'Q32', 'Q36', 'Q37', 'Q40', 'Q41'
]
```

```python
# Create df_old_trait_data DataFrame
data_list = []
for item in df_survey['state_trait_data']:
    if item == 'No Data':
        data_list.append({}) #empty case
    else:
        data_list.append(item) #non-empty case
df_old_trait_data = pd.DataFrame(data_list)

df_old_trait_data = df_old_trait_data.applymap(lambda x: x[0], na_action = 'ignore' )  #remove list
df_old_trait_data = pd.concat([df_survey[['userid']],df_old_trait_data],axis=1) #add userid
#inverse answer
for column in REVERSE_QUESTIONS_OLD:
    if column in df_old_trait_data.columns:
        df_old_trait_data[column] = 5 - df_old_trait_data[column]
#calculate trait_score and state_score
df_old_trait_data['trait_score'] = df_old_trait_data.loc[:, 'Q2':'Q21'].sum(axis=1)
df_old_trait_data['state_score'] = df_old_trait_data.loc[:, 'Q22':'Q41'].sum(axis=1)
df_old_trait_data = df_old_trait_data[['userid','trait_score','state_score']] #filter columns

df_old_trait_data
```

```python
# Create df_new_trait_data DataFrame
data_list = []
for item in df_survey['new_statetrait_data']:
    if item == 'No Data':
        data_list.append({}) #empty case
    else:
        data_list.append(item) #non-empty case
df_new_trait_data = pd.DataFrame(data_list)

df_new_trait_data = df_new_trait_data.applymap(lambda x: x[0], na_action = 'ignore' )  #remove list
df_new_trait_data = pd.concat([df_survey[['userid']],df_new_trait_data],axis=1) #add userid
#inverse answer
for column in REVERSE_QUESTIONS_NEW:
    if column in df_new_trait_data.columns:
        df_new_trait_data[column] = 5 - df_new_trait_data[column]
#calculate trait_score and state_score
df_new_trait_data['trait_score'] = df_new_trait_data.loc[:, 'Q2':'Q21'].sum(axis=1)
df_new_trait_data['state_score'] = df_new_trait_data.loc[:, 'Q22':'Q41'].sum(axis=1)
df_new_trait_data = df_new_trait_data[['userid','trait_score','state_score']] #filter columns

df_new_trait_data
```

```python
#merge both old and new into one

df_final_trait_data = pd.merge(df_old_trait_data, df_new_trait_data, on='userid', suffixes=('_old', '_new'))
# Update 'trait_score' and 'state_score' based on the conditions
df_final_trait_data['trait_score'] = df_final_trait_data.apply(lambda row: row['trait_score_new'] if row['trait_score_old'] == 0 else row['trait_score_old'], axis=1)
df_final_trait_data['state_score'] = df_final_trait_data.apply(lambda row: row['state_score_new'] if row['state_score_old'] == 0 else row['state_score_old'], axis=1)
df_final_trait_data.drop(['trait_score_old', 'trait_score_new', 'state_score_old', 'state_score_new'], axis=1, inplace=True)# Drop columns

#change 0 value to null
df_final_trait_data['state_score'].replace(0, np.nan, inplace=True)
df_final_trait_data['trait_score'].replace(0, np.nan, inplace=True)

df_final_trait_data
```

**Calculate Trust_score**

```python
REVERSE_QUESTIONS_TRUST = [
    'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q8', 'Q10', 'Q11', 'Q12', 'Q14', 'Q16', 'Q20', 'Q25'
]
```

```python
df_trust = pd.DataFrame(df_survey['trust_scale_data'].to_list())
df_trust = df_trust.applymap(lambda x: x[0])  #remove list
df_trust = pd.concat([df_survey[['userid']],df_trust],axis=1) #add userid
#inverse answer
for column in REVERSE_QUESTIONS_TRUST:
    if column in df_trust.columns:
        df_trust[column] = 8 - df_trust[column]
df_trust['trust_score'] = df_trust.loc[:, 'Q2':'Q26'].sum(axis=1) #calculate trust_score
df_trust = df_trust[['userid','trust_score']] #filter column

df_trust
```

**Merge all df**

```python
df_score = pd.merge(df_demographic,df_final_trait_data, on='userid', how='inner')
df_score = pd.merge(df_score,df_trust, on='userid', how='inner')

df_score
```
